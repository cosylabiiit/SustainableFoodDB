```python
ing_cat = pd.merge(cat,ing,on='RecipeDB Ingredient', how='right')

# If an ingredient is present in 'ing' but category is not present, fill with NaN
ing_cat.loc[ing_cat['Dietrx_Category'].isnull(), 'Dietrx_Category'] = pd.NA

ing_cat = ing_cat.drop(columns={'generic_name'})
ing_cat
```

```python
sueat = pd.read_csv("D:\IIITD\Mansi_Thesis_Research\Foodprint\Carbon_Footprint_July23\Data\Food_Items_CF_clean_1.csv")
sueat = sueat.rename(columns={'FOOD COMMODITY ITEM updated':'Sueatable Ingredient'})
# sueat['Sueatable Ingredient'] = sueat['Sueatable Ingredient'].apply(lambda x: x.title())

sueat = sueat.iloc[:, :6]
sueat 
```

```python
df1 = pd.merge(ing_cat,sueat, on='Sueatable Ingredient')
df1 = df1.drop(columns={'FOOD COMMODITY TYPOLOGY', 'FOOD COMMODITY sub-TYPOLOGY'})
df1
```

```python
sueatf = pd.read_csv("D:\IIITD\Mansi_Thesis_Research\Foodprint\Carbon_Footprint_July23\Data\Sueatable_preprocessed_data.csv")
```

```python
sueatf = sueatf.drop(columns={'FOOD COMMODITY GROUP','FOOD COMMODITY TYPOLOGY'})
sueatf 
```

```python
def custom_title_case(value):
    exceptions = ["EU", "UK", "USA", "US"]
    words = value.split()
    words = [word if word.upper() in exceptions else word.title() for word in words]
    return ' '.join(words)

# Apply title case with exceptions to the 'Region' column
sueatf['Region'] = sueatf['Region'].apply(custom_title_case)
sueatf['Country'] = sueatf['Country'].apply(custom_title_case)
```

```python
# sueatf['Region'] = sueatf['Region'].apply(lambda x: x.title())
# sueatf['Country'] = sueatf['Country'].apply(lambda x: x.title())
sueatf['Type of source'] = sueatf['Type of source'].apply(lambda x: x.title())

sueatf
```

```python
# Group by 'Sueatable Ingredient' and aggregate other information
grouped = sueatf.groupby('FOOD COMMODITY ITEM').agg({
    'Region': lambda x: x.unique().tolist(),
    'Country': lambda x: x.unique().tolist(),
    'Full Reference': lambda x: x.unique().tolist(),
    'Publication year': lambda x: x.unique().tolist(),
    'Type of source': lambda x: x.unique().tolist()
}).reset_index()

grouped
```

```python
final = pd.merge(df1,grouped,on='FOOD COMMODITY ITEM', how = 'left')
final = final.drop(columns={'FOOD COMMODITY ITEM'})
final = final.rename(columns={'Dietrx_Category':'Category', 'FOOD COMMODITY GROUP':'Food Commodity Group','FOOD COMMODITY TYPOLOGY updated':'Food Commodity Typology', 'Publication year':'Publication Year', 'CF':'Carbon Footprint', 'Type of source':'Source Type'})
final
```

```python
# Reorder columns
column_order = ['RecipeDB Ingredient', 'Category', 'Sueatable Ingredient', 'Food Commodity Group',
    'Food Commodity Typology', 'Region', 'Country', 'Carbon Footprint', 'Full Reference', 'Publication Year', 'Source Type']

final = final[column_order]

final
```

```python
#title case 

final['RecipeDB Ingredient'] = final['RecipeDB Ingredient'].apply(lambda x: x.title())
final['Sueatable Ingredient'] = final['Sueatable Ingredient'].apply(lambda x: x.title())
final['Food Commodity Group'] = final['Food Commodity Group'].apply(lambda x: x.title())
final['Food Commodity Typology'] = final['Food Commodity Typology'].apply(lambda x: x.title())
# final['Region'] = final['Region'].apply(lambda x: [item.title() for item in x])
# final['Country'] = final['Country'].apply(lambda x: [item.title() for item in x])
# final['Source Type'] = final['Source Type'].apply(lambda x: [item.title() for item in x])

final
```
