```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = [10, 20, 50, 100, 200]
models = ['BERT', 'RoBERTa', 'Word2Vec']

acc_b = [90, 85, 78, 77, 81.5]  # BERT
acc_r = [90, 70, 64, 61, 61]    # RoBERTa
acc_w = [70, 60, 52, 55, 50]    # Word2Vec

f_b = [93.61, 90.86, 84.5, 83.91, 86.32]
f_r = [93.61, 81.24, 74.99, 71.93, 70.88]
f_w = [79.51, 70.53, 55.55, 57.14, 54.45]

# Calculate bar width and positions
bar_width = 0.25
index = np.arange(len(x))

# Create the figure and axis for accuracy
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Plot the accuracy bars for each group
ax1.bar(index, acc_b, bar_width, color='blue', label='BERT')
ax1.bar(index + bar_width, acc_r, bar_width, color='cornflowerblue', label='RoBERTa')
ax1.bar(index + 2 * bar_width, acc_w, bar_width, color='lightsteelblue', label='Word2Vec')
ax1.set_xlabel('Top X (most popular) ingredients')
ax1.set_ylabel('Accuracy')
ax1.set_xticks(index + bar_width)
ax1.set_xticklabels(x)
ax1.legend()

# Plot the line plot for F-scores
ax2.plot(x, f_b, marker='o', color='blue', label='BERT')
ax2.plot(x, f_r, marker='o', color='cornflowerblue', label='RoBERTa')
ax2.plot(x, f_w, marker='o', color='lightsteelblue', label='Word2Vec')
ax2.set_xlabel('Top X (most popular) ingredients')
ax2.set_ylabel('F1-Score')
ax2.legend()

plt.tight_layout()
plt.show()
```
