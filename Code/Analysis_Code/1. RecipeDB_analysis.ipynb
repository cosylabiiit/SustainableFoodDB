# Frequency Rank Distrubution

```python
import ast
import operator
count = {}

recipes = recipe["ID"].nunique()

for i in range(recipes):
    r = recipe['ingredients'][i]
    res = ast.literal_eval(r)
    for k in res:
        count[k] = count.get(k,0) + 1
sorted_d = dict(sorted(count.items(), key=operator.itemgetter(1),reverse=True))

y=[]
for i in sorted_d.values():
    y.append(i)
    
x=[]
for i in range(20279):
    x.append(i)
    
plt.loglog(x,y,base=10)
plt.gcf().set_size_inches(8,6)
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.show()
```
