# carbon footprint size (values) distribution for 26 regions 

CF bins and number of recipes in each bins based on region

```python
import pandas as pd

# Define the bin size and range
bin_size = 5
cf_range = range(0, 121, bin_size)  # Bins of width 5 starting from 0 up to 120

# Create an empty list to store the bin counts
bin_counts_list = []

# Iterate over the CF bins
for i in range(len(cf_range) - 1):
    # Get the bin range
    bin_start = cf_range[i]
    bin_end = cf_range[i + 1]
    bin_label = f"{bin_start}-{bin_end}"

    # Create a dictionary to store the bin counts for each region
    bin_counts = {'CF_Bin': bin_label}

    # Iterate over unique regions
    for region in recipe['region'].unique():
        region_data = recipe[(recipe['region'] == region) & (recipe['Recipe_CF'] >= bin_start) & (recipe['Recipe_CF'] < bin_end)]
        count = len(region_data)
        bin_counts[region] = count

    # Append the bin counts to the list
    bin_counts_list.append(bin_counts)

# Create the DataFrame from the list of dictionaries
bin_counts_df = pd.DataFrame(bin_counts_list)

# Sort the DataFrame columns based on ['CF_Bin'] + region_carbon_fp_df['Region'].unique()
sorted_columns = ['CF_Bin'] + list(region_carbon_fp_df['Region'].unique())
bin_counts_df = bin_counts_df.reindex(columns=sorted_columns)

bin_counts_df
```

sorted based on median cf

# CF size distribution with world

```python
import pandas as pd

# Define the bin size and range
bin_size = 5
cf_range = range(0, 121, bin_size)  # Bins of width 5 starting from 0 up to 120

# Create an empty list to store the bin counts
bin_counts_list = []

# Iterate over the CF bins
for i in range(len(cf_range) - 1):
    # Get the bin range
    bin_start = cf_range[i]
    bin_end = cf_range[i + 1]
    bin_label = f"{bin_start}-{bin_end}"

    # Create a dictionary to store the bin counts for each region
    bin_counts = {'CF_Bin': bin_label}

    # Iterate over unique regions
    for region in recipe['region'].unique():
        region_data = recipe[(recipe['region'] == region) & (recipe['Recipe_CF'] >= bin_start) & (recipe['Recipe_CF'] < bin_end)]
        count = len(region_data)
        bin_counts[region] = count
    
    # Calculate world-level bin count (considering the entire dataset)
    world_data = recipe[(recipe['Recipe_CF'] >= bin_start) & (recipe['Recipe_CF'] < bin_end)]
    world_count = len(world_data)
    bin_counts['World'] = world_count
    
    # Append the bin counts to the list
    bin_counts_list.append(bin_counts)

# Create the DataFrame from the list of dictionaries
bin_counts_world_df = pd.DataFrame(bin_counts_list)

# Sort the DataFrame columns based on ['CF_Bin'] + region_carbon_fp_df['Region'].unique()
sorted_columns = ['CF_Bin'] + list(world_region_cf_df['Region'].unique())
bin_counts_world_df = bin_counts_world_df.reindex(columns=sorted_columns)

bin_counts_world_df
```
