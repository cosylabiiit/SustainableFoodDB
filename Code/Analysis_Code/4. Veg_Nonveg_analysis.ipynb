**We created a table that contains information on the total number of recipes, the number of vegetarian recipes, the number of non-vegetarian recipes, the number of miscellaneous recipes for each country.**


```python
import pandas as pd

region_data = []

total_count = mergedb_10[mergedb_10['Vegetarian_Recipe']==1].groupby('region')['Vegetarian_Recipe'].sum()
veg_count = mergedb_10[mergedb_10['Vegetarian_Recipe']==1].groupby('region')['Vegetarian_Recipe'].sum()
non_veg_count = mergedb_10[mergedb_10['Non_Vegetarian_Recipe']==1].groupby('region')['Non_Vegetarian_Recipe'].sum()
misc_count = mergedb_10[mergedb_10['Miscellaneous_Recipe']==1].groupby('region')['Miscellaneous_Recipe'].sum()

# Iterate over unique regions
for region in mergedb_10['region'].unique():
    veg = mergedb_10[mergedb_10['region']==region]['Vegetarian_Recipe']
    
    region_data_dict = {
    'region': region,
    'vegetarian': veg_count[region],
    'non_vegetarian': non_veg_count[region],
    'miscellaneous': misc_count[region],
    'count':len(veg.values.tolist())
    # 'total_cf': sum(region_data)
    }
    region_data.append(region_data_dict)


# Create a DataFrame from the list of dictionaries
region_veg_nonveg_df = pd.DataFrame(region_data)

# Print the DataFrame
region_veg_nonveg_df
```

# Statistical significance of CF value of veg and non-veg recipes

apply T-Test

The t-statistic (t_stat) measures the difference in means between the two groups relative to the variation within each group.
The p-value (p_value) represents the probability of observing the data under the assumption that there is no true difference in means between the two groups.

To determine the statistical significance:

If the p-value is less than your chosen significance level (e.g., 0.05), you can reject the null hypothesis and conclude that there is a statistically significant difference in CF between vegetarian and non-vegetarian recipes.
If the p-value is greater than your significance level, you fail to reject the null hypothesis, indicating that there is insufficient evidence to conclude a significant difference in CF between the two groups.

equal_var = “False”: The Welch’s t-test will be conducted by not taking into consideration the equal population variances.

```python
veg_recipes = mergedb_10[mergedb_10['Vegetarian_Recipe'] == 1]
cf_veg = veg_recipes['Recipe_CF']

non_veg_recipes = mergedb_10[mergedb_10['Non_Vegetarian_Recipe'] == 1]
cf_nonveg = non_veg_recipes['Recipe_CF']
```

```python
import scipy.stats as stats

t_stat, p_value = stats.ttest_ind( list(cf_nonveg),list(cf_veg), equal_var=False)

if p_value < 0.05:
    print("There is a statistically significant difference in CF between vegetarian and non-vegetarian recipes.")
else:
    print("There is no statistically significant difference in CF between vegetarian and non-vegetarian recipes.")
    
print(t_stat)
print(p_value)
```

```python
import scipy.stats as stats

# Perform the Kolmogorov-Smirnov test
statistic, p_value = stats.ks_2samp(cf_veg, cf_nonveg)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant difference in CF distributions between vegetarian and non-vegetarian recipes.")
else:
    print("Fail to reject the null hypothesis. There is no statistically significant difference in CF distributions between vegetarian and non-vegetarian recipes.")

print("KS statistic:", statistic)
print("p-value:", p_value)
```

p-value= 0 indicates an extremely strong statistical evidence against the null hypothesis and suggests that the observed difference between the groups is highly significant.

```python
import matplotlib.pyplot as plt
import numpy as np

regions = world_region_cf_veg_nonveg['region']
p_values = []
t_stats = []
significance = []

for i in range(len(regions)):
    t_stat, p_value = stats.ttest_ind(world_region_cf_veg_nonveg['cf_veg'][i], world_region_cf_veg_nonveg['cf_nonveg'][i])
    p_values.append(p_value)
    t_stats.append(t_stat)
    if p_value < 0.05:
        significance.append("Yes")
    else:
        significance.append("No")

fig, ax = plt.subplots(figsize=(10, 6))

x = np.arange(len(regions))
bar_width = 0.4

# Plot the p-values
ax.bar(x, p_values, width=bar_width, label='p-value')
ax.axhline(y=0.05, color='red', linestyle='--', label='Significance Threshold')
ax.set_xticks(x)
ax.set_xticklabels(regions, rotation=45, ha='right')

# Plot the t-statistics as markers
ax.scatter(x, t_stats, color='blue', marker='o', label='t-statistic')

# Add labels and title
ax.set_xlabel('Region')
ax.set_ylabel('Value')
ax.legend()

# Add significance labels
for i in range(len(regions)):
    ax.text(x[i], p_values[i] + 0.01, significance[i], ha='center')

plt.tight_layout()
plt.show()
```
