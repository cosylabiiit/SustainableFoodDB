# Calculate cf for each recipe after string matching

```python
# create a dictionary to map ingredients to their carbon footprints
ingredient_dict = dict(zip(df_result['RecipeDB Ingredient'].str.lower(), df_result['CF']))
# print(ingredient_dict)

# create a function to compute the sum of carbon footprints for a list of ingredients
def compute_sum(ingredients):
    ingredients = [x.strip(" '[]") for x in ingredients.split(",")]
    
    # map the ingredients to their carbon footprints using the dictionary
    carbon_footprints = []
    for ing in ingredients:
        if ing.lower() in ingredient_dict:
            carbon_footprints.append(ingredient_dict[ing.lower()])
        else:
            carbon_footprints.append(0)

    return sum(carbon_footprints)

df_sum = pd.DataFrame({'Ingredients': input_file['Available Ingredients'], 
                       'carbon_footprint_sum': input_file['Available Ingredients'].apply(compute_sum)})
df_sum
```

```python
input_file['Recipe_CF'] = input_file['Available Ingredients'].apply(compute_sum)
input_file
```
